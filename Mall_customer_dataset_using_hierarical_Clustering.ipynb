```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt


# Agglomerative Clustering
model = AgglomerativeClustering(
    n_clusters=4,
    metric='euclidean',
    linkage='ward'
)

labels = model.fit_predict(a)

print("Cluster Labels:")
print(labels)

# Visualization
plt.scatter(a[:,0], a[:,1], c=labels, cmap='rainbow')
plt.title("Agglomerative Clustering")
plt.show()






```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


def divisive_clustering(data, n_clusters):
    clusters = [data]

    while len(clusters) < n_clusters:
        largest_cluster = max(clusters, key=len)
        clusters.remove(largest_cluster)

        kmeans = KMeans(n_clusters=2, random_state=42)
        labels = kmeans.fit_predict(largest_cluster)

        cluster1 = largest_cluster[labels == 0]
        cluster2 = largest_cluster[labels == 1]

        clusters.append(cluster1)
        clusters.append(cluster2)

    return clusters

clusters = divisive_clustering(a, 3)
for i, cluster in enumerate(clusters):
    print(f"\nCluster {i+1}")
    print(cluster)

# Visualization
colors = ['red', 'blue', 'green', 'purple']

for i, cluster in enumerate(clusters):
    plt.scatter(cluster[:,0], cluster[:,1],
                color=colors[i % len(colors)],
                label=f'Cluster {i+1}')

plt.legend()
plt.title("Divisive Clustering")
plt.show()
```
