```python
from sklearn.impute import SimpleImputer
im=SimpleImputer(strategy='most_frequent')
d[['Age']]=im.fit_transform(d[['Age']])
from sklearn.impute import SimpleImputer
im=SimpleImputer(strategy='most_frequent')
d[['Age']]=im.fit_transform(d[['Cabin']])
```

```python
d=d.drop('Cabin',axis=1)
```

```python
#Categorical encoding
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
d['Sex']=l.fit_transform(d['Sex'])
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
d['Embarked']=l.fit_transform(d['Embarked'])
d
```

```python
x=d[['SibSp','Parch']]
y=d[["Survived"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Fitting the dataset
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(criterion='entropy',verbose=1)
clf=clf.fit(x_train,y_train)
```

```python
#predicting
y_predict=clf.predict(x_test)
```

```python
#Heatmap
import seaborn as sns
correlation_matrix=x.corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix,cmap='coolwarm',annot=True)
plt.show()
```

```python
#Evaluatin metrics
from sklearn.metrics import accuracy_score
print("Accuracy score: ",accuracy_score(y_test,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
print(cm)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_predict)
print(cr)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Pick one tree from the forest
tree = clf.estimators_[0]

# Plot the tree
plt.figure(figsize=(20,10))

plot_tree(
    tree,
    feature_names=x_train.columns,
    class_names=[str(c) for c in clf.classes_],
    filled=True,
    rounded=True,
    fontsize=8
)

plt.title("Random Forest - Tree 1")
plt.show()
```
