```python
from transformers import pipeline

summarizer = pipeline("summarization")

text = """
Artificial Intelligence (AI) is one of the most important technologies of the modern era.
It enables machines to perform tasks that normally require human intelligence, such as
understanding language, recognizing images, solving problems, and making decisions.
AI is widely used in healthcare, education, finance, agriculture, and transportation.
In healthcare, AI helps doctors diagnose diseases and recommend treatments.
In education, it supports personalized learning experiences for students.
In finance, AI is used for fraud detection and risk analysis.
Agriculture benefits from AI through smart irrigation systems, crop monitoring, and yield prediction.
Transportation systems use AI for traffic management and self-driving vehicles.
As AI continues to develop, it is expected to improve productivity, reduce costs, and create new opportunities.
However, there are also challenges such as data privacy, security, and ethical concerns that need to be addressed.
Researchers and organizations around the world are working to ensure that AI is developed responsibly and used for the benefit of society.
**Artificial Intelligence and Its Impact on Modern Society (≈600 words)**

Artificial Intelligence (AI) has become one of the most transformative technologies of
the modern era, influencing nearly every aspect of human life and reshaping the way individuals,
 businesses, and governments operate. AI refers to the ability of machines and computer systems to
  perform tasks that traditionally require human intelligence, such as learning, reasoning,

"""
summary = summarizer(text)

print(summary)
```

```python
from transformers import pipeline

translator = pipeline(
    "translation_en_to_fr",
    model="Helsinki-NLP/opus-mt-en-fr"
)

result = translator("Hello, how are you?")

print(result)
```

```python
from transformers import pipeline

q= pipeline("question-answering")

context = """
Artificial Intelligence (AI) is a technology that enables machines
to perform tasks that normally require human intelligence.
"""

question = "What does AI enable machines to do?"

r = q(
    question=question,
    context=context
)

print(r["answer"])
```
