```python
#Categorical encoding
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
d['Geography']=l.fit_transform(d['Geography'])
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
d['Gender']=l.fit_transform(d['Gender'])
d
```

```python
d=d.drop('Surname',axis=1)
```

```python
x=d[["CreditScore","Tenure","Balance","IsActiveMember","EstimatedSalary"]]
#x=d.drop('Exited',axis=1)
y = d[["Exited"]]
```

```python
#Standard scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scale= sc.fit_transform(d)
print(scale)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Fitting the dataset
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf=clf.fit(x_train,y_train)
```

```python
#Standard scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scale= sc.fit_transform(d)
print(scale)
```

```python
#predicting
y_predict=clf.predict(x_test)
```

```python
#Evaluatin metrics
from sklearn.metrics import accuracy_score
print("Accuracy score: ",accuracy_score(y_test,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
print(cm)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_predict)
print(cr)
```
