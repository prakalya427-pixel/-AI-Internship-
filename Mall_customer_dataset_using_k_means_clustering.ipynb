```python
a=d.iloc[:,3:].values
a
```

```python
from sklearn.cluster import KMeans
c=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,random_state=0)
  kmeans.fit(a)
  c.append(kmeans.inertia_)
plt.plot(c)
```

```python
kmeans=KMeans(n_clusters=5,init=('k-means++'),random_state=42)
y=kmeans.fit_predict(a)
```

```python
kmeans.inertia_
```

```python
kmeans.labels_
```

```python
plt.scatter(a[y==0,0],a[y==0,1],color='green',label='cluster1')
plt.scatter(a[y==1,0],a[y==1,1],color='blue',label='cluster2')
plt.scatter(a[y==2,0],a[y==2,1],color='red',label='cluster3')
plt.scatter(a[y==3,0],a[y==3,1],color='yellow',label='cluster4')
plt.scatter(a[y==4,0],a[y==4,1],color='pink',label='cluster5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],
            color='black',label='centroids',s=300,marker=',')
```
