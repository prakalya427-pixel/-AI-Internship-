```python
#Categorical encoding
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
d['policy_state']=l.fit_transform(d['policy_state'])
d
d['authorities_contacted']=l.fit_transform(d['authorities_contacted'])
d
d['police_report_available']=l.fit_transform(d['police_report_available'])
d
d['fraud_reported']=l.fit_transform(d['fraud_reported'])
d
d['auto_model']=l.fit_transform(d['auto_model'])
d
```

```python
from sklearn.impute import SimpleImputer
im=SimpleImputer(strategy='most_frequent')
d[['authorities_contacted']]=im.fit_transform(d[['authorities_contacted']])
print(d[['authorities_contacted']])
d.isnull().sum()
```

```python
#Standard scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scale= sc.fit_transform(d.select_dtypes(include=np.number))
print(scale)
```

```python
#Heatmap
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix=d.corr(numeric_only=True)
print(correlation_matrix)
sns.heatmap(correlation_matrix,cmap='coolwarm',annot=True)
plt.show()
```

```python
x=d.drop('\tpolice_report_available',axis=1)
y = d[["\tpolice_report_available"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Fitting the dataset
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='entropy',max_depth=3)
clf=clf.fit(x_train,y_train)
```

```python
#predicting
y_predict=clf.predict(x_test)
```

```python
#Evaluatin metrics
from sklearn.metrics import accuracy_score
print("Accuracy score: ",accuracy_score(y_test,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
print(cm)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_predict)
print(cr)
```

```python
#tree visualization
import graphviz
from sklearn import tree
dot_data = tree.export_graphviz(
    clf,
    out_file=None,
    feature_names=x.columns,
    class_names=['No', 'Yes'],
    filled=True,
    rounded=True,
    special_characters=True
)
graph = graphviz.Source(dot_data)
graph
```
